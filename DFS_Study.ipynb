```python
dy = [1, 0, -1, 0]
dx = [0, 1, 0, -1]

def solution1(maze, queries):
    answer = []
    
    R = len(maze)
    C = len(maze[0])
    
    for q in queries:
        sp = q.split(" ")
        s = list(map(int, sp[:2]))
        e = list(map(int, sp[2:4]))
        al = set(sp[4])
        
        alb = dict()
        
        for a in al:
            alb[a] = True
            
        s = (s[0]-1, s[1]-1)
        e = (e[0]-1, e[1]-1)
        
        dq = [(s, 1)]
        visited = [[False] * C for _ in range(R)]
        
        posb = 10e4
        while dq:
            now, l = dq.pop(0)
            visited[now[0]][now[1]] = True
            
            if now == e:
                posb = min(posb, l)
                break
                
            for i in range(4):
                ny = now[0] + dy[i]
                nx = now[1] + dx[i]
                
                if 0 <= ny < R and 0 <= nx < C:
                    if alb.get(maze[ny][nx]) and not visited[ny][nx]:
                        if posb > l+1:
                            dq.append(((ny, nx), l+1))
                            
        if posb == 10e4:
            posb = -1
        answer.append(posb)
        
    return answer
```

```python
maze = [
    "AAAAA",
    "AABBB",
    "CAEFG",
    "AAEFF"
]
queries = [
    "1 1 1 5 AF",
    "1 1 4 5 AF",
    "2 1 4 5 FAE",
    "1 5 4 5 ABF",
    "1 1 4 1 A"
]

import time

start = time.time()
print(solution1(maze, queries))
end = time.time()
print(f"{end - start:.5f} sec")

start = time.time()
print(solution2(maze, queries))
end = time.time()
print(f"{end - start:.5f} sec")
```
